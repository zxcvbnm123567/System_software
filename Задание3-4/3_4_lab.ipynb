# Агрегирование данных с сайта 

```python
comission_district = {
    1: 'Адмиралтейский',
    2: 'Василеостровский',
    3: 'Кировский',
    4: 'Красногвардейский',
    5: 'Невский',
    6: 'Красносельский',
    7: 'Кировский',
    8: 'Петродворцовый',
    9: 'Приморский',
    10: 'Выборгский',
    11: 'Калининский',
    12: 'Приморский',
    13: 'Курортный',
    14: 'Выборгский',
    15: 'Кронштадский',
    16: 'Центральный',
    17: 'Калининский',
    18: 'Петроградский',
    19: 'Московский',
    20: 'Пушкинский',
    21: 'Колпинский',
    22: 'Выборгский',
    23: 'Фрунзенский',
    24: 'Невский',
    25: 'Красногвардейский',
    26: 'Красносельский',
    27: 'Московский',
    28: 'Приморский',
    29: 'Фрунзенский',
    30: 'Центральный'
}

def add_yavka(df:pd.DataFrame):
    yavka = {}
    yavka['-'] = 'Явка избирателей'
    for i, col in enumerate(df.columns[1:]):
        yavka[col] = round((int(df.iloc[7,i+1]) + int(df.iloc[8,i+1])) / int(df.iloc[0,i+1]) * 100, 2)
    first_part = pd.concat((df.loc[:11], pd.DataFrame(yavka, index=[0])), ignore_index=True)
    return pd.concat((first_part, df.loc[12:]), ignore_index=True)

def add_VIKs(df:pd.DataFrame):
    VIKs = {}
    for i in range(1,31):
        tik_name = r'comissions/tik_{}.htm'.format(i)
        soup_tik = BeautifulSoup(open(tik_name, encoding='utf-8'), "html.parser")
        VIKs[f'Территориальная избирательная комиссия №{i}'] = pd.read_html(str(soup_tik.find_all('table')[6]))[2].shape[1]
    VIKs['Цифровые избирательные участки'] = 1
    VIKs['Сумма'] = sum(VIKs.values())
    VIKs['-'] = 'Количество участковых комиссий на районе'
    
    first_part = pd.concat((df.loc[:10], pd.DataFrame(VIKs, index=[0])), ignore_index=True)
    return pd.concat((first_part, df.loc[11:]), ignore_index=True)

def comissions_to_district(df:pd.DataFrame, comission_district:dict)->pd.DataFrame:
    resulted_dict = {}
    resulted_dict['-'] = df['-'].values
    resulted_dict['Сумма'] = df['Сумма'].values
    
    for tik in comission_district:
        if tik not in resulted_dict:
            resulted_dict[comission_district[tik]] = df[f'Территориальная избирательная комиссия №{tik}'].values
        else:
            temp_arr = resulted_dict[comission_district[tik]][:11] + df[f'Территориальная избирательная комиссия №{tik}'].values[:11]
            
            temp_arr = np.append(temp_arr, None)
            
            for i in range(3):
                split = resulted_dict[comission_district[tik]][12+i].split(' ')
                val_1 = int(split[0])
                
                split_2 = df[f'Территориальная избирательная комиссия №{tik}'].values[12+i].split(' ')
                val_2 = int(split_2[0])
                
                temp_arr = np.append(temp_arr, f'{val_1+val_2} {(val_1+val_2)/sum(temp_arr[7:9])}%')
            
            resulted_dict[comission_district[tik]] = temp_arr
    resulted_dict['Цифровые избирательные участки'] = df['Цифровые избирательные участки']
    resulted_df = pd.DataFrame(resulted_dict)
    return resulted_df
```

```python
# Данные взяты с сайта http://www.st-petersburg.vybory.izbirkom.ru
soup = BeautifulSoup(open(r"downloaded_site.htm", encoding='utf-8'), "html.parser")

part1_df = pd.read_html(str(soup.find_all('table')[6].find_all('table')[0]))[0]
part2_df = pd.read_html(str(soup.find_all('table')[6].find_all('table')[1]))[0]

columns = [*['-', 'Сумма'], *part2_df.iloc[0].values]

part1_vals = part1_df[[1,2]].iloc[1:].values.tolist()

for i, row in enumerate(part1_vals):
    part1_vals[i] = [*row, *part2_df.iloc[1:].values[i].tolist()]
all_values = part1_vals

aggregated_df = pd.DataFrame(data=all_values, columns=columns)
aggregated_with_VIKs_df = add_VIKs(df=aggregated_df)
district_df = comissions_to_district(df=aggregated_with_VIKs_df, comission_district=comission_district)
district_with_yavka_df = add_yavka(district_df)
district_with_yavka_df.to_csv('district_with_yavka_df.csv', index=False)
district_with_yavka_df.head(3)
```

```python
from bs4 import BeautifulSoup
from shapely.geometry import Polygon, MultiPolygon
from mpl_toolkits.axes_grid1 import make_axes_locatable
from geopandas import GeoSeries
import requests
from tqdm import tqdm
import geopandas as gpd
```

```python
def get_polygon(name)->Polygon:
    soup = BeautifulSoup(open(r"maps_xml/{}.xml".format(name), encoding='utf-8'), "html.parser")
    coords = []
    for way in tqdm(soup.find_all('member')):

        xml_way = f"https://www.openstreetmap.org/api/0.6/way/{way['ref']}"
        response = requests.get(xml_way).text
        way_soup = BeautifulSoup(response)

        for node in way_soup.find_all('nd'):
            xml_node = f"https://www.openstreetmap.org/api/0.6/node/{node['ref']}"
            response_node = requests.get(xml_node).text
            node_soup = BeautifulSoup(response_node)
            lat = node_soup.find('node')['lat']
            lon = node_soup.find('node')['lon']
            coords.append((float(lon), float(lat)))
    return Polygon(coords)

polygon_list = []
for col in district_with_yavka_df.columns[2:-1]:
    polygon_list.append(get_polygon(col))
```
